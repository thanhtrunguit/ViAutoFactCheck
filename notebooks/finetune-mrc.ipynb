```python
def compute_answer_start(row):
    ctx = row['context']
    ans = row['evidence']
    if pd.isna(ans) or str(ans).strip() == "":
        return None
    ans = str(ans).strip()
    start = ctx.find(ans)
    return int(start) if start != -1 else None

train["answer_start"] = train.apply(compute_answer_start, axis=1)

# Convert to Int64 dtype and filter to rows that DO contain the evidence
train["answer_start"] = pd.to_numeric(train["answer_start"], errors="coerce").astype("Int64")
num_missing = train["answer_start"].isna().sum()
print(f"Total rows: {len(train)}, rows without evidence in context (will be skipped): {num_missing}")
```

```python
def preprocess_function(examples):
    questions = [q.strip() for q in examples["claim"]]
    inputs = tokenizer(
        questions,
        examples["context"],
        max_length=512,
        truncation="only_second",
        return_offsets_mapping=True,
        padding="max_length",
    )

    offset_mapping = inputs.pop("offset_mapping")
    answers = examples["answers"]
    start_positions = []
    end_positions = []

    for i, offset in enumerate(offset_mapping):
        answer = answers[i]

        # Nếu answer_start bị thiếu hoặc rỗng, gán (0, 0)
        if not answer["answer_start"] or pd.isna(answer["answer_start"][0]):
            start_positions.append(0)
            end_positions.append(0)
            continue

        start_char = answer["answer_start"][0]
        end_char = start_char + len(answer["text"][0])
        sequence_ids = inputs.sequence_ids(i)

        # Find the start and end of the context
        idx = 0
        while sequence_ids[idx] != 1:
            idx += 1
        context_start = idx
        while idx < len(sequence_ids) and sequence_ids[idx] == 1:
            idx += 1
        context_end = idx - 1

        # If the answer is not fully inside the context, label it (0, 0)
        if offset[context_start][0] > end_char or offset[context_end][1] < start_char:
            start_positions.append(0)
            end_positions.append(0)
        else:
            # Otherwise it's the start and end token positions
            idx = context_start
            while idx <= context_end and offset[idx][0] <= start_char:
                idx += 1
            start_positions.append(idx - 1)

            idx = context_end
            while idx >= context_start and offset[idx][1] >= end_char:
                idx -= 1
            end_positions.append(idx + 1)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs
```

```python
import os
import re
import torch
from torch.utils.data import DataLoader
from tqdm import tqdm
from transformers import AutoTokenizer, AutoModelForQuestionAnswering
import pandas as pd

# ---------- HYPERPARAMS ----------
TEST_CSV_PATH = "test_vinum_formatted.csv"
TOP_K_MODELS = 2

# Sliding window parameters
WINDOW_SIZE = 480   # max tokens in each subcontext (reserve some tokens for claim + special tokens)
MIN_STRIDE = 200     # overlap stride
MAX_LENGTH = 512

# ---------- GLOBAL TEST DATA LOADING (load once) ----------
test_df = pd.read_csv(TEST_CSV_PATH)
print(len(test_df))
# ---------- UTILITIES ----------
import shutil

def safe_remove_dir(path):
    if path and os.path.exists(path):
        try:
            shutil.rmtree(path)
            print(f"Removed directory: {path}")
        except Exception as e:
            print(f"Warning: failed to remove {path}: {e}")


# ---------- UTILITIES ----------
def normalize_for_em(s):
    if s is None:
        s = ""
    s = str(s).strip().lower()
    s = re.sub(r"[^\w\s]", "", s)
    s = " ".join(s.split())
    return s

def exact_match(pred, gold):
    return int(normalize_for_em(pred) == normalize_for_em(gold))
    
def tok_len_pair(claim, context):
    """Return tokenizer length for claim + context input to model"""
    return tokenizer(claim, context, truncation=False, add_special_tokens=True, padding=False, return_length=True)["length"][0]

def context_ids_no_specials(text):
    return tokenizer(text, add_special_tokens=False)["input_ids"]

def decode_ids(ids):
    return tokenizer.decode(ids, skip_special_tokens=True, clean_up_tokenization_spaces=True)

def create_subcontexts_sliding_window(claim, context, max_len=MAX_LENGTH):
    if tok_len_pair(claim, context) <= max_len:
        return [context.strip()]

    ctx_ids = context_ids_no_specials(context)
    claim_tok_len = len(tokenizer(claim, add_special_tokens=False)["input_ids"])
    reserve = claim_tok_len + 3  # special tokens buffer
    window_size = min(WINDOW_SIZE, max_len - reserve)
    stride = max(MIN_STRIDE, int(window_size * 0.33))
    step = window_size - stride
    if step <= 0:
        step = max(1, window_size // 2)

    subcontexts = []
    for start in range(0, len(ctx_ids), step):
        end = start + window_size
        window_ids = ctx_ids[start:end]
        window_text = decode_ids(window_ids)

        # safety: trim right until claim + window fits
        while tok_len_pair(claim, window_text) > max_len and len(window_ids) > 0:
            window_ids = window_ids[:-1]
            window_text = decode_ids(window_ids)

        if window_text.strip():
            subcontexts.append(window_text.strip())
    return subcontexts


# ---------- ANSWERING FUNCTION ----------
def answer_question_local(model, tokenizer, device, question: str, context: str, max_length=MAX_LENGTH):
    encodings = tokenizer(
        question,
        context,
        return_tensors="pt",
        truncation="only_second",
        max_length=max_length,
        return_offsets_mapping=True,
        padding="max_length",
    )
    input_ids = encodings["input_ids"].to(device)
    attention_mask = encodings["attention_mask"].to(device)
    offset_mapping = encodings["offset_mapping"][0]

    with torch.no_grad():
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        start_logits = outputs.start_logits
        end_logits = outputs.end_logits

    start_idx = torch.argmax(start_logits, dim=1).item()
    end_idx = torch.argmax(end_logits, dim=1).item()
    if end_idx < start_idx:
        end_idx = start_idx

    start_char, _ = offset_mapping[start_idx].tolist()
    _, end_char = offset_mapping[end_idx].tolist()
    return context[start_char:end_char].strip()

# ---------- COMPUTE EM (no saving, use preloaded test_df) ----------
def compute_em_on_test(model, tokenizer, device):
    """
    Evaluates EM over the global test_df (same as before), but uses the tokenizer-aware
    sliding-window splitting when (claim, context) exceed MAX_LENGTH. Accepts a predicted
    answer only if it has more than 10 words (as you requested).
    """
    em_total = 0
    n = len(test_df)

    for _, row in tqdm(test_df.iterrows(), total=n, desc="Evaluating EM"):
        question = str(row["claim"])
        context = str(row["context"])
        gold_evidence = row.get("evidence", "")

        pred_evidence = ""

        total_len = tok_len_pair(question, context)


        if int(total_len) <= MAX_LENGTH:
            pred_evidence = answer_question_local(model, tokenizer, device, question, context)
        else:
            # exceeds — split using token-level sliding windows and run sequentially
            segments = create_subcontexts_sliding_window(question, context, max_len=MAX_LENGTH)
            # Run model over subcontexts, accept first substantive answer (> 10 words)
            for seg in segments:
                if not seg or not seg.strip():
                    continue
                try:
                    seg_pred = answer_question_local(model, tokenizer, device, question, seg)
                except Exception:
                    seg_pred = ""

                if seg_pred and seg_pred.strip():
                    # accept only if prediction has more than 10 words
                    if len(seg_pred.strip().split()) > 10:
                        pred_evidence = seg_pred
                        break
                    # else, continue to next segment (reject too-short predictions)

        # compute EM incrementally
        gold_text = "" if pd.isna(gold_evidence) else str(gold_evidence)
        em_total += exact_match(pred_evidence, gold_text)

    em_score = em_total / n if n > 0 else 0.0
    print(f"\nExact Match (EM): {em_score:.4f} ({int(em_total)}/{n})")
    return em_score

# ---------- TRAINING LOOP ----------
def train_qa(
    model,
    tokenizer,
    processed_squad,
    device,
    epochs=5,
    batch_size=8,
    lr=2e-5,
    max_length=MAX_LENGTH,
    top_k=TOP_K_MODELS,
    save_dir_root="best_models"
):
    processed_squad["train"].set_format(
        type="torch",
        columns=["input_ids", "attention_mask", "start_positions", "end_positions"]
    )
    train_loader = DataLoader(processed_squad["train"], batch_size=batch_size, shuffle=True)
    optimizer = torch.optim.AdamW(model.parameters(), lr=lr)
    top_models = []

    for epoch in range(epochs):
        
        # ---- TRAINING PHASE ----
        
        model.train()
        running_loss = 0.0

        for batch in tqdm(train_loader, desc=f"Training Epoch {epoch+1}"):
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            start_positions = batch["start_positions"].to(device)
            end_positions = batch["end_positions"].to(device)

            optimizer.zero_grad()
            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                start_positions=start_positions,
                end_positions=end_positions,
            )
            loss = outputs.loss
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        avg_loss = running_loss / len(train_loader)
        print(f"\nEpoch {epoch+1}/{epochs} - Training Loss: {avg_loss:.4f}")

        # ---- TESTING PHASE ----
        model.eval()
        
        print("→ Evaluating EM on test set ...")
        val_em = compute_em_on_test(model, tokenizer, device)

        save_path = os.path.join(save_dir_root, f"epoch{epoch+1}_em{val_em:.4f}")

        if len(top_models) < top_k or val_em > min(top_models, key=lambda x: x[0])[0]:
            # save new candidate
            save_path = os.path.join(save_dir_root, f"epoch{epoch+1}_em{val_em:.4f}")
            os.makedirs(save_path, exist_ok=True)
            model.save_pretrained(save_path)
            tokenizer.save_pretrained(save_path)
            print(f"Saved candidate model to {save_path}")

            # add to list and sort
            top_models.append((val_em, epoch + 1, save_path))
            top_models = sorted(top_models, key=lambda x: x[0], reverse=True)

            # identify dropped models (beyond top_k)
            if len(top_models) > top_k:
                dropped = top_models[top_k:]
                top_models = top_models[:top_k]
                for em_s, ep, pth in dropped:
                    safe_remove_dir(pth)

            print(f"Saved model (epoch {epoch+1}) with EM={val_em:.4f}")

        print("Current top models:")
        for em_s, ep, pth in top_models:
            print(f"  EM={em_s:.4f} | epoch={ep} | {pth}")

    print("Training finished.")
    return top_models
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

top_models = train_qa(
    model=model,
    tokenizer=tokenizer,
    processed_squad=processed_squad,
    device=device,
    epochs=10,
    batch_size=16,
    lr=2e-6,
    max_length=MAX_LENGTH,
    top_k=2,
    save_dir_root="saved_top_em_models"
)
```
